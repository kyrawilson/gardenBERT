```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../probe_results.csv")
```

```python
gp_u = df.loc[(df['code'].str.match('GP-U'))]
gp_a = df.loc[(df['code'].str.match('GP-A'))]
```

```python
# Subselect GPs
gp_u = df.loc[(df['code'].str.match('GP-U'))]
gp_a = df.loc[(df['code'].str.match('GP-A'))]

# Garden path means
gp_u_n1_arg0_mean = gp_u['n1_arg0_1_prob'].mean()
gp_u_n1_arg1_mean = gp_u['n1_arg1_1_prob'].mean()

gp_u_n2_arg0_mean = gp_u['n2_arg0_1_prob'].mean()
gp_u_n2_arg1_mean = gp_u['n2_arg1_1_prob'].mean()


print("Garden Path Results - Unambiguous")

print(f"Agent Matrix arg0 is {gp_u_n1_arg0_mean} and arg1 is {gp_u_n1_arg1_mean} ")
print(f"Patient Matrix arg0 is {gp_u_n2_arg0_mean} and arg1 is {gp_u_n2_arg1_mean} ")


gp_a_n1_arg0_mean = gp_a['n1_arg0_1_prob'].mean()
gp_a_n1_arg1_mean = gp_a['n1_arg1_1_prob'].mean()

gp_a_n2_arg0_mean = gp_a['n2_arg0_1_prob'].mean()
gp_a_n2_arg1_mean = gp_a['n2_arg1_1_prob'].mean()

print("Garden Path Results - - Ambiguous")

print(f"Agent Matrix arg0 is {gp_a_n1_arg0_mean} and arg1 is {gp_a_n1_arg1_mean} ")
print(f"Patient Matrix arg0 is {gp_a_n2_arg0_mean} and arg1 is {gp_a_n2_arg1_mean} ")

```

```python
# Plot GP Results
# Theme
arg0_color = "#BB5566"
arg1_color = "#DDAA33"

# Data
gp_a_res = [gp_a_n1_arg0_mean, gp_a_n1_arg1_mean, gp_a_n2_arg0_mean, gp_a_n2_arg1_mean]
gp_labels = ["First Noun, Arg0", 'First Noun, ARG1', "Secound Noun, Arg0", "Second Noun, Arg1"]

sns.set_theme(style="whitegrid")
plot = sns.barplot( gp_labels, gp_a_res, palette=[arg0_color, arg1_color])
plot.set_ylim(0, 1)
for label in (plot.get_yticklabels()):
	label.set_fontsize(20)


plt.savefig("plots/proberesults/gp-a.png", format="png", transparent=True) # specify filetype explicitly
```
